```python
df["LSTAT"].fillna(method='ffill',inplace=True)
df["CRIM"].fillna(method='ffill',inplace=True)
df["ZN"].fillna(method='ffill',inplace=True)
df["INDUS"].fillna(method='ffill',inplace=True)
df["CHAS"].fillna(method='ffill',inplace=True)
df["AGE"].fillna(method='ffill',inplace=True)
```

```python
df.isnull().sum()
```

**Define independent and dependent variable from dataset**

```python
#first semi colon take all rows and second semi colon take columns
x=df.iloc[:,0:14]  #select all rows and columns from 0 to 12
y=df.iloc[:,-1]    #select all rows and columns from 
```

**Splitting data into training and testing dataset**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

**Shape of training and testing data**

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

**Standardizing the data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

**Importing linaer regression function**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X_train,Y_train)
y_pred=lr.predict(X_test)
```

```python
print(y_pred)
```

**Root means square root**

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(Y_test,y_pred))
```

```python
print("Root means squared is: ",rmse)
```

**Training accuracy**

```python
print("Training Accuracy is: ",lr.score(X_train,Y_train))
```

```python
print("Testing Accuracy is: ",lr.score(X_test, Y_test))
```
